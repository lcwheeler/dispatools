#### Demonstration of plotting multiple 1D spectra and polar plots simultaneously

```python
from dispatools import parse_dataset, panel_plot
```

```python
# Use the parse_dataset() function to import all of the example shoulder spectra
test2 = parse_dataset("../example_data/Shoulder/")
```

```python
# Use the panel_plot function to create a panel of 1D spectra and matched polar plots
ascii_peaks_ref = "../example_data/ascii-spec.txt"
panel_plot(test2, "test2-panel-shoulders", ascii_peaks_ref, threshold = 0.05, units_polar="a.u.", units_1d="Hz", figsize=(8, 18), color="midnightblue")
```

### Demonstration of overlay plotting - showing all polar plots color-coded on the same graph

```python
from dispatools import parse_dataset, overlay_plot
```

```python
# Use the parse_dataset() function to load the example dataset from specified directory
op = "../example_data/OnePeak/"
tol = parse_dataset(op)
```

```python
# Create a list of colors to pass to the plotting function
#colors = ["darkcyan","darkorange","indigo","violet","lime","gold","blue","fuchsia","darkred","dodgerblue"]
colors = ["darkcyan","darkorange","indigo","violet"]
```

```python
# dictionary of phase shifts for labeling the subplots
#labels = {'3000':r"0$^\circ$", '4000':r"0$^\circ$", '4045':r"45$^\circ$", '4090':r"90$^\circ$", 
                       #'4135':r"135$^\circ$", '4180':r"180$^\circ$", '4225':r"225$^\circ$", '4270':r"270$^\circ$", '4315':r"315$^\circ$"}

labels = {'1000':r"0$^\circ$", '1090':r"90$^\circ$", '1180':r"180$^\circ$", '1270':r"270$^\circ$"}

labels
```

```python
overlay_plot(tol, "test_overlay_plot",  "../example_data/ascii-spec.txt", colors = colors, labels = labels, 
                              threshold = 0.05, units_polar="a.u.", units_1d="Hz", figsize=(8, 5))
```
